```python
import librosa
import scipy as sp
import matplotlib.pyplot as plt
import IPython.display as ipd
import numpy as np
```

```python
audio_path = "/kaggle/input/datasets/garvs777/libri3mix/Libri3Mix/wav8k/min/train-100/mix_clean/103-1240-0001_3879-174923-0064_1594-135914-0012.wav"
audio_path_s1 = "/kaggle/input/datasets/garvs777/libri3mix/Libri3Mix/wav8k/min/train-100/s1/103-1240-0001_3879-174923-0064_1594-135914-0012.wav"
audio_path_s2 = "/kaggle/input/datasets/garvs777/libri3mix/Libri3Mix/wav8k/min/train-100/s2/103-1240-0001_3879-174923-0064_1594-135914-0012.wav"
audio_path_s3 = "/kaggle/input/datasets/garvs777/libri3mix/Libri3Mix/wav8k/min/train-100/s3/103-1240-0001_3879-174923-0064_1594-135914-0012.wav"
ipd.Audio(audio_path)
```

```python
signal , sr = librosa.load(audio_path)
signal_s1 , sr_s1 = librosa.load(audio_path_s1)
signal_s2 , sr_s2 = librosa.load(audio_path_s2)
signal_s3 , sr_s3 = librosa.load(audio_path_s3)
plt.figure(figsize=(18,8))
time = np.linspace(0, len(signal) / sr, num=len(signal))
plt.plot(time, signal, color="blue", alpha = 0.5)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(18,8))
time = np.linspace(0, len(signal) / sr, num=len(signal))
plt.plot(time, signal_s1, color="green",alpha=0.7)
plt.plot(time, signal_s2, color="yellow", alpha=0.7)
plt.plot(time, signal_s3, color="red", alpha=0.4)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.tight_layout()
plt.show()
```

```python
ft = sp.fft.fft(signal)
magnitude = np.absolute(ft)
frequency = np.linspace(0,sr, len(magnitude))
```

```python
plt.figure(figsize=(18,8))
plt.plot(frequency[:5000],magnitude[:5000])
plt.xlabel("Frequency")
plt.ylabel("Magnitude")
plt.show()
```
